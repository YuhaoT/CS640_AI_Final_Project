```python
import pandas as pd
import json
import shutil
```

```python
labeled_users = './data/labeled_users.csv'
user_demo_profiles_path = './data/User demo profiles.json'

df_users = pd.read_csv(labeled_users)
df_profiles = pd.read_json(user_demo_profiles_path, orient='values')
```

```python
df_profiles
```

```python
df_tmp = df_profiles.drop(['name','screen_name','description','lang'],axis = 1)

for index, row in df_tmp.iterrows():
    id = row[0]
    path = row[1]
    source_path = './data/img' + path[12:]
    dest_path = './data/' + path

    try:
        shutil.copyfile(source_path,dest_path)
    except(FileNotFoundError):
        print(id,path)
```
