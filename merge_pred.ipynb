```python
import pandas as pd
import numpy as np
import torch 
from sklearn.metrics import classification_report
```

```python
img_pred = pd.read_csv('./race_prediction_image/userid_race_predrace.csv')
```

```python
tweet_pred = pd.read_csv('./race_prediction_image/tweets_race_pred_prob.csv')
```

```python
img_pred['0'].value_counts()
```

```python
img_pred.columns
```

```python
tweet_pred.head()
```

```python
race_merged = pd.merge(img_pred, tweet_pred, on='user_id')
```

```python
race_merged.dropna(inplace=True)
race_merged
```

```python
img_prob0 = race_merged['0'].to_numpy()
img_prob1 = race_merged['1'].to_numpy()
img_prob2 = race_merged['2'].to_numpy()
img_prob3 = race_merged['3'].to_numpy()

tweet_prob0 = race_merged['pred_prob_0'].to_numpy()
tweet_prob1 = race_merged['pred_prob_1'].to_numpy()
tweet_prob2 = race_merged['pred_prob_2'].to_numpy()
tweet_prob3 = race_merged['pred_prob_3'].to_numpy()
```

```python
img_prob = np.vstack((img_prob0, img_prob1, img_prob2, img_prob3)).T
print(img_prob.shape)

tweet_prob = np.vstack((tweet_prob0, tweet_prob1, tweet_prob2, tweet_prob3)).T
tweet_prob.shape
```

```python
from sklearn import preprocessing
norm_img_prob = preprocessing.normalize(img_prob)
print(norm_img_prob)

norm_tweet_prob = preprocessing.normalize(tweet_prob)
norm_tweet_prob
```

#### Race Prediction based on probability returned by image

```python
preds = np.argmax(img_prob, axis = 1)
print(preds)
print(classification_report(race_merged['label'], preds))
```

#### Race Prediction based on probability returned by tweer

```python
preds = np.argmax(tweet_prob, axis = 1)
print(preds)
print(classification_report(race_merged['label'], preds))
```

#### Race Prediction based on probability returned by image and bert model

```python
probs = tweet_prob + img_prob
preds = np.argmax(probs, axis = 1)
print(preds)
print(classification_report(race_merged['label'], preds))
```

```python
age_merged = pd.merge(img_age_pred, tweet_age_pred, on='screen_name')
```

```python
age_merged
```

```python
img_prob0 = age_merged['P0'].to_numpy()
img_prob1 = age_merged['P1'].to_numpy()
tweet_prob0 = 1 + age_merged['pred_prob_0'].to_numpy()
tweet_prob1 = 1 + age_merged['pred_prob_1'].to_numpy()
```

```python
img_prob = np.vstack((img_prob0, img_prob1)).T
tweet_prob = np.vstack((tweet_prob0, tweet_prob1)).T
```

```python
norm_tweet_prob = preprocessing.normalize(tweet_prob)
norm_img_prob = preprocessing.normalize(img_prob)
```

#### Prediction based on probability returned by image

```python
preds = np.argmax(img_prob, axis = 1)
print(classification_report(age_merged['label_y'], preds))
```

#### Prediction based on probability returned by tweets

```python
preds = np.argmax(tweet_prob, axis = 1)
print(classification_report(age_merged['label_y'], preds))
```

#### Prediction based on probability returned by image and bert model

```python
probs = (tweet_prob * 0.7) + (img_prob * 0.3)
preds = np.argmax(probs, axis = 1)
print(classification_report(age_merged['label_y'], preds))
```
